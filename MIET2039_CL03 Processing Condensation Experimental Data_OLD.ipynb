Boundary layer thickness:

$$ \delta (x) = \left [ \frac{4k_l \mu_l \left( T_{sat}-T_s \right )x}{g\rho_l \left (\rho_l-\rho_v \right ) h_{fg}}\right ]^{1/4} $$

Velocity profile in the film:

$$ u_y = \frac{\left( \rho_L - \rho_G \right )}g{\mu_L}\left[  y \delta - \frac{y^2}{2} \right ] $$

$$ h = 0.943\left( \frac{\rho_L\left( \rho_L - \rho_g\right)gh_{fg}k_L^3}{4\mu_Lk_L\left( T_{sat}-T_s\right)}\right) $$

```python
rhoL = 998     #density of water
rhoG = 1.225   #density of air
rhoV = 0.7     #density of vapour

muL = 8.9e-4   #dynamic viscosity of water
g = 9.81       # gravitational acceleration

kL = 0.6      #conductivity of water
hfg = 2314000 #latent heat of vaporization J/kg

Tsat = 100    #saturation temperature
Ts = 80       #surface temperature

import numpy as np
x = np.array([i/800 for i in range(51)])
#print (x)
bl = ((4*kL*muL*(Tsat - Ts)*x)/(g*rhoL*(rhoL-rhoV)*hfg))**(1/4)

import matplotlib.pyplot as plt
myPlot = plt.plot(bl,x, label='boundary layer thickness')
plt.ylabel('distance from top of plate')
plt.xlabel('boundary layer thickness')
plt.legend(loc='upper left', shadow=True)
```

```python
rhoL = 990     #density of water
rhoG = 0.601   #density of air

muL = 375e-6   #dynamic viscosity of water
g = 9.81       # gravitational acceleration
blThick = 0.0001 #boundary layer thickness (m)

import numpy as np
y = np.array([i*blThick/100 for i in range(101)])
u_y = ((rhoL - rhoG)*g/muL)*(y*blThick - y*y/2)
```

```python
import matplotlib.pyplot as plt
myPlot = plt.plot(y, u_y, label='velocity profile')
plt.xlabel('y (m)')
plt.ylabel('velocity (m/s)')
plt.legend(loc='upper left', shadow=True)
```
